#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
# TODO
casa_cara = ds.loc[ds['price'].idxmax()]
                   
print(f"La casa con dirección en {casa_cara['address']} es la casa más cara y su precio es de {casa_cara['price']} ")
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
# TODO
ds_filtrado = ds[ds['price'] > 0]
casa_cheap = ds_filtrado.loc[ds_filtrado['price'].idxmin()]
                   
print(f"La casa con dirección en {casa_cheap['address']} es la casa más barata y su precio es de {casa_cheap['price']} ")
```

#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

Imprime la dirección y el área de las casas seleccionadas. Por ejemplo:

`La casa más grande está ubicada en Calle Gran Vía, Nº38 y su superficie es de 5000 metros`

`La casa más pequeña está ubicada en Calle Mayor, Nº12 y su superficie es de 200 metros`

Este ejercicio es similar al anterior, pero buscamos las casas más grandes y pequeñas en base a su superficie.

```python
# TODO
casa_big = ds.loc[ds['surface'].idxmax()]
print(f"La casa más grande está ubicada en {casa_big['address']} y su suyperficie es de {casa_big['surface']} ")

casa_small = ds.loc[ds['surface'].idxmin()]
print(f"La casa más pequeña está ubicada en {casa_small['address']} y su suyperficie es de {casa_small['surface']} ")
```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

```python
# TODO
media_arroyomolinos = ds[ds['level5'] == 'Arroyomolinos (Madrid)']
precio_arroyomolinos = media_arroyomolinos['price'].mean()
print(precio_arroyomolinos)


```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt

# Trazar el histograma de precios

precio_arroyomolinos = ds[ds['level5'] == 'Arroyomolinos']['price'].dropna()

plt.hist(precio_arroyomolinos, color='red')
plt.title('Histograma de precios Arroyomolinos')
plt.xlabel('Precio')
plt.ylabel('Población')
plt.show()

```
